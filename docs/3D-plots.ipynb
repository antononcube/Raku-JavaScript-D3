## 3D random points

```raku
my @generators =
    {&random-variate(NormalDistribution.new(1, 1), $_)}, 
    {&random-variate(NormalDistribution.new(1, 1.4), $_)}, 
    {&random-variate(NormalDistribution.new(4, 3), $_)},
;

my @data = random-tabular-dataset(300, <x y z>, :@generators);

deduce-type(@data)
```

```raku
sink records-summary(@data)
```

```raku
#%js
js-d3-list-point-plot3d(@data, :$background, :!axes, :boxed)
```

## 3D random points with groups

```raku
my @generators =
    {&random-variate(NormalDistribution.new(1, 1), $_)}, 
    {&random-variate(NormalDistribution.new(1, 1.4), $_)}, 
    {&random-variate(NormalDistribution.new(4, 3), $_)},
    { ('a' xx $_).List }
;

my @data1 = random-tabular-dataset(500, <x y z group>, :@generators);

@generators =
    {&random-variate(NormalDistribution.new(5, 1), $_)}, 
    {&random-variate(NormalDistribution.new(7, 3.4), $_)}, 
    {&random-variate(NormalDistribution.new(6, 2), $_)},
    { ('b' xx $_).List }
;

my @data2 = random-tabular-dataset(500, <x y z group>, :@generators);

my @data = |@data1, |@data2;

deduce-type(@data)
```

```raku
sink records-summary(@data)
```

```raku
#%js
my $box-ratios = [1, 1, 0.4];
js-d3-list-point-plot3d(
    @data, 
    :$title-color,
    x-label => 'X',
    x-label-color => 'red',
    x-label-font-size => 16,
    y-label => 'Y',
    y-label-color => 'blue',
    z-label => 'Z',
    z-label-color => 'green',
    :$background, 
    :$box-ratios, 
    :axes,
    :!boxed,
    :legends,
    :1200width,
    :700height
)
```

```raku
#%js
js-d3-list-point-plot3d(
    @data, 
    :$background, 
    box-ratios => Whatever, 
    color-scheme => 'Category10',
    :!axes, 
    :boxed, 
    :ticks, 
    title => 'Random points', 
    :!legends
)
```

## 3D line plot

```raku
my @data = (1...4).map( -> $r { (0, π/30 ... 2*π + π/30).map(-> $t { <x y z group>.Array Z=> [$r * sin($t), $r * cos($t), cos($r*$t), $r.Str] })».Hash })».List;
@data .= map(*.Slip);
deduce-type(@data)
```

```raku
#%js
my @color-scheme = "#a4a74e", "#f28e2b", "#e15759", "#76b7b2", "#59a14f", "#edc948", "#b07aa1", "#ff9da7", "#9c755f", "#bab0ab";
js-d3-list-line-plot3d(
    @data,
    :$background,
    :@color-scheme
)
```
